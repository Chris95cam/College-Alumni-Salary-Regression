## Data Processing, Part 1

### Loading Data

I created a web scraper using BeautifulSoup to take data from forbes.com and save it to a csv file. The source contains information for 650 colleges across the United States. The columns in the dataset are:
1. Rank - College ranking by Forbes Magazine; based on alumni salary, student satisfaction, student debt, American leaders, on-time graduation rate, and academic success
2. Name - Name of the college
3. City & State - Location of the college campus
4. Public/Private - Funding type of the school
5. Undergraduate Population - Number of enrolled undergraduate students
6. Student Population -Total number of students enrolled 
7. Net Price - Average cost for one year of education subtracting any financial aid received by its students
8. Average Grant Aid - Average amount of money students receive each year to help pay tuition, from sources such as the government or the university itself
9. Total Annual Cost - Total cost of tuition, room and board, and any additional fees that the college charges per year
10. Alumni Salary - Median salary for workers with 10 or more years of experience (payscale.com)
11. Acceptance Rate - Percentage of students who apply to a college that are admitted
12. SAT Composite Range - First and third quartile composite SAT scores of incoming freshman
13. ACT Composite Range - First and third quartile composite ACT scores of incoming freshman
14. Website - URL for college's website

```python
#load in data collected from web scrapper
df = pd.read_csv(r'C:\Users\CCCam\OneDrive\DataScience\ForbesTopUniversities2019\ForbesAmericasTopColleges2019Original.csv')
df.head()
```

### Cleaning Data

The csv file contains raw data gathered from the website. To clean this data, we need to accomplish three main things:
1. Extract the city and state from the 'City & State' column
2. Remove punctuation from numerical columns and convert them to an appropriate data type
3. Extract upper and lower values from the test score columns

```python
#extract City and State from 'City & State' (some rows have only state information)
df['City'] = df['City & State'].apply(lambda x:  x.split(', ')[0] if len(x.split(', '))==2 else np.NaN) 
df['State'] = df['City & State'].apply(lambda x: x.split(', ')[-1])
df.drop(columns='City & State', inplace =True)

#remove punctuation from numerical columns and convert to float
for column in ['Rank','Undergraduate Population','Student Population','Total Annual Cost','Alumni Salary', 'Net Price',
               'Average Grant Aid','Acceptance Rate']:
    for char in ['#',',','$','%']:
        df[column] = df[column].str.replace(char,'')
    df[column] = df[column].astype(float)

#extract upper and lower test score values
df[['SAT Lower', 'SAT Upper']] = df['SAT Composite Range'].str.split(expand=True,pat='-').astype(float)
df[['ACT Lower', 'ACT Upper']] = df['ACT Composite Range'].str.split(expand=True,pat='-').astype(float)

df.drop(columns = ['SAT Composite Range', 'ACT Composite Range'], inplace =True)
```

Next, we reorganize the columns in the dataframe and save the cleaned version to a new csv file. 

```python
#reorganize columns
df = df[['Rank','Name','City','State','Public/Private','Undergraduate Population', 'Student Population', 'Net Price',
       'Average Grant Aid', 'Total Annual Cost','Alumni Salary','Acceptance Rate','SAT Lower', 'SAT Upper', 'ACT Lower', 'ACT Upper','Website']]

#save to csv file
df.to_csv('ForbesAmericasTopColleges2019.csv',index=False)

df.head()
```

### Removing Unnecessary Data

Some information in the dataframe will not be needed during the rest of the project. This includes:
1. Website - It doesn't make sense that information from a college's website url can help predict the salary of its alumni
2. Rank - The ranking methodology Forbes uses factors in alumni salary, so it would not make sense to predict earnings from this feature. Also, many different college ranking lists exist, so by removing this column, the model can generalize to schools that are not in this list
3. Rows with alumni salary missing - This is the target feature, so any row where this information is not present should be removed

```python
#drop columns
df.drop(columns = ['Website','Rank'],inplace = True)

#drop rows with null values (all missing values belong to 'Alumni Salary' columns)
df = df.dropna(subset=['Alumni Salary'])

#show new column makeup
df.info()
```

Now, we will group the data by state and funding type, and impute values for 'Net Price', 'Average Grant Aid', and 'Acceptance Rate' based on the means of these groups.

There are many missing values for SAT and ACT scores (almost 100), so we will take more care with imputing these values. We will generate a feature correlation matrix to see which features are correlated with these test scores.

```python
#plot heatmap of numerical feature correlations
plt.figure(figsize=(8,8))
sns.heatmap(df.corr().round(2),annot=True)
plt.title('Numerical Feature Correlations',fontweight='bold')
plt.show()
```

Average test scores of a college seem to be strongly correlated with acceptance rate. Test scores are also highly correlated with alumni salary, but it would not make sense to impute test scores based on this as it is the feature we are trying to predict. There is also a moderately strong correlation between test scores and average grant aid as well as total annual cost. However, average grant aid, total annual cost, and acceptance rate all have moderately strong correlations with each other, so to avoid multicollinearity, we will impute based only on acceptance rate, which has the strongest correlation with test scores. Let's make a chart of just test scores and acceptance rate.


```python
#set up figure
fig,axes = plt.subplots(2,sharex=True,figsize=(9,6))
plt.suptitle('Test Scores and Acceptance Rate',fontweight='bold',y=.9)
sns.set_style('whitegrid')

#plot SAT scores
sns.regplot(data=df,x='Acceptance Rate',y ='SAT Upper',scatter_kws={'alpha':0.4},ax=axes[0])
sns.regplot(data=df,x='Acceptance Rate',y ='SAT Lower',scatter_kws={'alpha':0.4},ax=axes[0])
axes[0].set_ylabel('SAT Scores')
axes[0].legend(['Upper','Lower'])

#plot ACT scores
sns.regplot(data=df,x='Acceptance Rate',y ='ACT Upper',scatter_kws={'alpha':0.4},ax=axes[1])
sns.regplot(data=df,x='Acceptance Rate',y ='ACT Lower',scatter_kws={'alpha':0.4},ax=axes[1])
axes[1].set_ylabel('ACT Scores')

plt.show()
```

The following code uses linear regression to impute the NaN values. Inside the loop, a linear regression model is fit for each test score, and predictions are made for each NaN entry. We can also see that all values are now present in the dataframe.

```python
for score in ['SAT Lower', 'SAT Upper', 'ACT Lower', 'ACT Upper']:
    #drop null values
    df_impute = df[['Acceptance Rate', score]].dropna()
    
    #create independant and dependant variable sets
    X = df_impute['Acceptance Rate'].values.reshape(-1,1)
    y = df_impute[score]
    
    #fit model
    LR = LinearRegression()
    LR.fit(X,y)
    
    #predict missing values
    missing = df['Acceptance Rate'][df[score].isnull()]
    predicted = LR.predict(missing.values.reshape(-1,1))
    
    #fill missing values (found method to fill NaN values online, see acknowledgements)
    fill = missing.replace(to_replace=missing.values,value=predicted)
    df[score].fillna(value=fill,inplace=True)
    
df.info()
```

### Feature Engineering

Now that the data has been cleaned, we will create new features that may have have an impact on predicting alumni salary. In this section, the three objectives are to:
1. Take the middle values of the test score ranges. This will provide a more accurate representation of the typical score for each test.

2. Get the ratio of middle ACT score to middle SAT score for each college. This will show how much a school weighs one test over the other in the admissions process.

3. Find the proportion of undergraduate students to total students. This will show how much of a focus each college puts on its undergraduates. 

4. Categorize the type of college based on its name. Words such as 'University', and 'Institute' appear frequently in the names of colleges, so we will see if these designations have any impact on alumni earnings. 

```python
#middle value of test score ranges
df['SAT Middle'] = (df['SAT Lower']+df['SAT Upper'])/2
df['ACT Middle'] = (df['ACT Lower']+df['ACT Upper'])/2

#'weighting' of ACT vs SAT
df['ACT/SAT Weight'] = (df['ACT Middle']/df['SAT Middle'])

#calculate proportion of undergrads to total students
df['Undergrad_prop'] = df['Undergraduate Population']/df['Student Population']

#categorize type of school based on name
df['University'] = df['Name'].apply(lambda x: True if x.find('University')!= -1 else False)
df['College'] = df['Name'].apply(lambda x: True if x.find('College')!= -1 else False)
df['Institute'] = df['Name'].apply(lambda x: True if x.find('Institute')!= -1 else False)
df['School'] = df['Name'].apply(lambda x: True if x.find('School')!= -1 else False)
df['Academy'] = df['Name'].apply(lambda x: True if x.find('Academy')!= -1 else False)
df.drop(columns=['Name'],inplace = True) #now that we have extracted info from name of college, we will drop it 

#view summary statistics of newly created features
df.describe()
```

## Data Processing, Part 2

### Features and Label

The next step is to prepare the data to be fit to machine learning models. We will start by splitting the features and target label in to separate variables, while also calling get_dummies in to transform the categorical data columns to numerical data columns.

```python
#create feature dataframe
X = df.drop(columns='Alumni Salary')
X = pd.get_dummies(X)

#create target label series
y = df['Alumni Salary']

X.head()
```

### Standardization

From the exploratory data analysis performed above, we can see that many of the features in the dataset are highly correlated with each other. To reduce multicollinearity, we will use principal component analysis on the numerical features in the 'X' dataframe. We will begin the process by standardizing these features. A heat map of feature correlations are shown below.

```python
#standardize numerical data columns
scaler = StandardScaler()
numerical = X.columns[:14]
X_numerical_std = pd.DataFrame(scaler.fit_transform(X[numerical]),columns=numerical)

#plot heatmap of numerical feature correlations
plt.figure(figsize=(8,8))
sns.heatmap(X_numerical_std.corr().round(2),annot=True)
plt.title('Numerical Feature Correlations',fontweight='bold')
plt.show()
```

### Principal Component Analysis

Next, to determine how many principal components to use, we will plot a line graph to show the explained variance ratio for each number of components. 

```python
#fit PCA for all numerical columns
pca = PCA(n_components=14)
pca.fit(X_numerical_std)

#plot explained variance ratio by number of components
plt.xlabel('Number of Components')
plt.ylabel('Explained Variance Ratio')
plt.title('Principal Component Analysis',fontweight='bold')
sns.lineplot(x=np.arange(1,15), y=pca.explained_variance_ratio_, color='black')
plt.show()
```

The 'elbow' in the chart above occurs at 3, so we will keep this number of components. We can observe the makeups of each component by creating a bar chart of feature contributions for each. Component 1 is made up of primarily test scores, acceptance rate, and cost. Component 2 is made up mostly from undergraduate population and student population. Component 3 is mostly influenced by undergraduate proportion.  

```python
#fit PCA for optimal number of components
pca = PCA(n_components=3)
pca.fit(X_numerical_std)

#plot make-ups of principal components
components = pca.components_
fig, axes = plt.subplots(ncols=3,sharey=True,figsize=(12,6))

#component 1
sns.barplot(x=components[0],y=numerical,ax=axes[0])
axes[0].set_xlim([-.8,.8])
axes[0].set_title('Component 1')

#component 2
sns.barplot(x=components[1],y=numerical,ax=axes[1])
axes[1].set_xlim([-.8,.8])
axes[1].set_title('Component 2')

#component 3
sns.barplot(x=components[2],y=numerical,ax=axes[2])
axes[2].set_xlim([-.8,.8])
axes[2].set_title('Component 3')

plt.suptitle('Feature Makeups of Principal Components',fontweight='bold')  
plt.show()
```

Next, we will re-create the dataframe using the principal components derived from above. We will accomplish this using the following steps:

1. Create a  new dataframe using the pca-transformed numerical features, and rename the principal components based on their feature makeups

2. Remove the old numerical features from the original dataframe, and store it as a new dataframe

```python
#rename principal components
X_pca = pd.DataFrame(pca.transform(X_numerical_std),
                     columns=['Barrier to Entry (PC 1)','Population Size (PC 2)',
                              'Undergrad Focus (PC 3)']).reset_index(drop=True)

#remove numerical features from dataframe
X_no_numerical = X.drop(columns=numerical).reset_index(drop=True)

#replace numerical features with principal components
X_reduced = pd.concat([X_pca, X_no_numerical],axis=1)

X_reduced.head()
```

Before we fit the data to machine learning models, we will create a benchmark that can be used to judge the effectiveness of these models. ### Feature Importances

To see the feature importances of the SVR, we will train the model over the entire data set. We will print out the optimal hyper-parameters from the grid-search.

```python
#print best hyperparameters from grid-search for SVR
print('Best Parameters:',SV.best_params_)
```

Using these hyper-parameters we will fit the model, and plot the feature coefficients. It appears that colleges with 'Institute' or 'Academy' in their name are the best contributers to alumni salary of the categorical features.

```python
#fit regressor over entire dataset
regressor = SVR(C=1000, kernel='linear')
regressor.fit(X_reduced ,y)

#function to plot feature importances of SVR (taken from webstie, see acknowledgements)
def plot_coefficients(classifier, feature_names, top_features=16):
    #order coefficients
    coef = classifier.coef_.ravel()
    top_positive_coefficients = np.argsort(coef)[-top_features:]
    top_negative_coefficients = np.argsort(coef)[:top_features]
    top_coefficients = np.hstack([top_negative_coefficients, top_positive_coefficients])
    
    # create plot
    plt.figure(figsize=(15, 5))
    colors = ['red' if c < 0 else 'blue' for c in coef[top_coefficients]]
    plt.bar(np.arange(2 * top_features), coef[top_coefficients], color=colors)
    feature_names = np.array(feature_names)
    plt.title('SVR Feature Coefficients (Most Important)',fontweight='bold')
    plt.xlabel('Feature')
    plt.ylabel('Coefficient')
    plt.xticks(np.arange(1, 1 + 2 * top_features), feature_names[top_coefficients], rotation=60, ha='right')
    plt.show()

#show plot
plot_coefficients(regressor, X_reduced.columns)
```
